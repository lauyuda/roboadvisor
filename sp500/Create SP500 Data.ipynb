```python
import pandas as pd
import yfinance as yf
import datetime
```

```python
tickers = pd.read_csv('sp500symbols.csv', encoding='latin-1')
```

```python
tickers.head()
```

```python
tickers_list = tickers['Symbol'].astype(str).values.tolist()
```

```python
start = datetime.datetime.now() - datetime.timedelta(days=3*365)
```

```python
end = datetime.datetime.now()
```

```python
tickers_list = [sub.replace('.', '-') for sub in tickers_list]
```

```python
tickers_list
```

```python
data = yf.download(tickers_list,
                   start=start, 
                   end=end,
                   progress=False,
                   auto_adjust=True)
```

```python
data.head()
```

```python
data = data['Close']
```
